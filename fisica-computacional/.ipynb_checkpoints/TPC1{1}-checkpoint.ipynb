## Pergunta 1
**Princípio de Huygens** - Christiaan Huygens (1629-1685) foi um físico, matemático e engenheiro Holandês que propôs um método de representação de frentes de onda, onde cada ponto de uma frente de onda se comporta como uma nova fonte de ondas elementares, que se propagam para além da região já atingida pela onda original e com a mesma frequência. Este ideia é conhecida por princípio de Huygens. Apesar de não ser exacta e ter vindo a sofrer sucessivas alterações (Fresnel, Kirchhoff), com o passar dos anos, ainda é ensinada pois permite introduzir alguns dos conceitos fundamentais da interferência de ondas.

Vamos considerar que um feixe laser é emitido de uma fonte colocada num ponto X com uma amplitude dada por um perfil gaussiano, 
\begin{equation}
U(x,y)=U_{0}e^{ikx}e^{-\frac{y^{2}}{2\sigma^{2}}},
\end{equation}
onde $k$ é número de onda em $xx'$ ($k=2\pi/\lambda$) e $\sigma$ é uma medida da largura do feixe. Neste problema iremos usar $\sigma=1$. Para aplicarmos o princípio de Huygens, vamos considerar una superfície $\Sigma$ que corta o feixe do laser. Podemos imaginar que cada ponto da superfície $(x_{s},y_{s})$, emite uma radiação circular (2D). Em qualquer ponto do espaço, a amplitude medida será dada pela sobreposição de todas as amplitudes das onduletas geradas pela superfície $\Sigma$ (a 2D é uma linha). A figura seguinte tem um esquema que representa esta construção.



Desta forma a amplitude secundária num ponto $\boldsymbol{r}=(x,y)$, é dada por 
\begin{equation}
U_{s}(x,y)=\frac{k}{2}\int_{\Sigma}dr_{s}U(x_{s},y_{s})H_{0}^{(1)}(\left|\boldsymbol{r}-\boldsymbol{r}_{s}\right|k)
\end{equation}
onde $H_{0}^{(1)}(x)$ é uma função de Hankel do primeiro tipo que pode ser chamada a partir do [https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.hankel1.html||scipy.special.hankel1] que resultou a integração do príncipio de Huygens em z para obter a teoria 2D. Considere $k=512$, $\sigma=1$ e $U_{0}=1$. Pretendemos calcular $U_{s}(10,0)$ considerando que a superfície $\Sigma$ é a linha infinita que passa em x=0. O integral é numérico dá (exacto a 25 casas decimais), 
\begin{equation}
U_{s}(10,0)=0.69247699620405105797219086532549115-0.72131043233017755094937790855952844i.
\end{equation}

```python
plt=reload(plt) # recomeça as definições de um plot para evitar possíveis erros

#----------------------------------------------------------------Funções-----------------------------------------------------------------------


# ---> Gaussiana <---
def gaussian(x, y, U0, k, sigma): #função gaussiana com nenhum parâmetro fixado
    expoente1 = (1j * k * x)
    expoente2 = -(y**2 / (2 * sigma **2))
    return U0 * np.exp(expoente1) * np.exp(expoente2)


# Durante todos os meus códigos chamarei xs,ys os ponto do plano x = 0
# e x,y os pontos com distância xp do plano x = 0


# ---> Integranda <---
def integranda(ys): # função integranda, k, U0 ,sigma, x,y e xs já definidos, variando apenas ys 
    
    (k, U0 ,sigma ,xs ,y ,x)= 512 ,1 ,1 , 0 , 0 , 10
    Us = gaussian(xs,ys,U0,k,sigma)
    
    return  (k/2)*(scipy.special.hankel1(0, k*np.sqrt((y - ys)**2 + (x - xs)**2),out = None)* Us)


# ---> Trapézio em uma variável <---
def trap (f,N,a,b): # método de integração do trapézio
    
    x = np.linspace(a,b,N+1)
    y = f(x)
    h = np.abs(b-a)/N
    integral = f(a)/2 + f(b)/2
    
    for i in range (1,N ):
        integral += f(a + i*h)
    return h * integral

# ---> Romberg <---
def romberg(f, Ni, grau, a, b, integral, e): # método de Romber com truncamento em um determinado erro

    N = grau # grau máximo da matriz de romberg
    R = np.zeros((N, N), complex)
    R[0][0] = integral(f, Ni, a, b)
    E = R[0][0]

    for i in range(1, N):
        Ni *= 2
        R[i][0] = integral(f, Ni, a, b)
        
        for k in range(1, i+1):  # Ajuste do intervalo do loop interno
            R[i][k] = R[i][k - 1] + (1 / (4 ** k - 1)) * (R[i][k - 1] - R[i - 1][k - 1])
            
        E = np.abs((1 / (4 ** (k + 1) - 1)) * (R[i][k] - R[i - 1][k]))  # Cálculo do erro de truncamento
        
        if E < e:  # Verifica se o erro é menor que o limite
            return R


#----------------------------------------------------------------Código-----------------------------------------------------------------------


ref = 0.69247699620405105797219086532549115-0.72131043233017755094937790855952844j

grau = 10 # grau exato que minha matriz deve ter, achado me maneira empírica
a = -5
b = 5

# Aplico meu Metodo de Romberg
R = romberg(integranda, 1, grau, a, b, trap, 10 ** (-6)) 

# Plotagem dos gráficos
for i in range(grau):
    for k in range(i,grau+1):
        n = 2 ** np.arange(0,k)
        val = R[:, i]
    erros = np.abs((val-ref)/ref)
    plt.plot(n[i:], erros[i:], label =f"Coluna {i+1}")


plt.xlabel('Número de intervalos')
plt.ylabel('log(Erro)')
plt.legend()
plt.yscale('log')
plt.grid(True)
plt.title('Erro em cada coluna')
plt.show()

#----------------------------------------------------------------Comentários----------------------------------------------------------------------------

# O que se observa é que quanto maior o número do grau de extrapolação mais rápido (com menos intervalos)
# O valor converge para o valor verdadeiro, o que já era esperado uma vez que cada coluna dom Romberg cai com
# uma potência maior de h , sendo h o tamanho de cada intervalo do limite de integração
```
