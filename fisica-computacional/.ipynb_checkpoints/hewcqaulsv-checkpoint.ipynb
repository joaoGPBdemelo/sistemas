## Pergunta 1
**Princípio de Huygens** - Christiaan Huygens (1629-1685) foi um físico, matemático e engenheiro Holandês que propôs um método de representação de frentes de onda, onde cada ponto de uma frente de onda se comporta como uma nova fonte de ondas elementares, que se propagam para além da região já atingida pela onda original e com a mesma frequência. Este ideia é conhecida por princípio de Huygens. Apesar de não ser exacta e ter vindo a sofrer sucessivas alterações (Fresnel, Kirchhoff), com o passar dos anos, ainda é ensinada pois permite introduzir alguns dos conceitos fundamentais da interferência de ondas.

Vamos considerar que um feixe laser é emitido de uma fonte colocada num ponto X com uma amplitude dada por um perfil gaussiano, 
\begin{equation}
U(x,y)=U_{0}e^{ikx}e^{-\frac{y^{2}}{2\sigma^{2}}},
\end{equation}
onde $k$ é número de onda em $xx'$ ($k=2\pi/\lambda$) e $\sigma$ é uma medida da largura do feixe. Neste problema iremos usar $\sigma=1$. Para aplicarmos o princípio de Huygens, vamos considerar una superfície $\Sigma$ que corta o feixe do laser. Podemos imaginar que cada ponto da superfície $(x_{s},y_{s})$, emite uma radiação circular (2D). Em qualquer ponto do espaço, a amplitude medida será dada pela sobreposição de todas as amplitudes das onduletas geradas pela superfície $\Sigma$ (a 2D é uma linha). A figura seguinte tem um esquema que representa esta construção.



Desta forma a amplitude secundária num ponto $\boldsymbol{r}=(x,y)$, é dada por 
\begin{equation}
U_{s}(x,y)=\frac{k}{2}\int_{\Sigma}dr_{s}U(x_{s},y_{s})H_{0}^{(1)}(\left|\boldsymbol{r}-\boldsymbol{r}_{s}\right|k)
\end{equation}
onde $H_{0}^{(1)}(x)$ é uma função de Hankel do primeiro tipo que pode ser chamada a partir do [https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.hankel1.html||scipy.special.hankel1] que resultou a integração do príncipio de Huygens em z para obter a teoria 2D. Considere $k=512$, $\sigma=1$ e $U_{0}=1$. Pretendemos calcular $U_{s}(10,0)$ considerando que a superfície $\Sigma$ é a linha infinita que passa em x=0. O integral é numérico dá (exacto a 25 casas decimais), 
\begin{equation}
U_{s}(10,0)=0.69247699620405105797219086532549115-0.72131043233017755094937790855952844i.
\end{equation}
1. Represente a função integranda no intervalo de $y_{s}$ de $[-15,15]$. Explique como sugere lidar com o limite de integração. 

##### O limite de integração será constante, entre y = -15 e y = 15, enquanto que o valor de x é fixo em x = 10 e o valor de y varia para cada novo momento de integração.

```python
# Dados do problema

k = 512 # vetor de onda
sigma = 1
U0 = 1
xs = 0

# Função U(x,y)

def U(x,y):
    return U0 * np.exp(1j * k *x) * np.exp(-(y**2)/(2*(sigma**2)))

# Função de Hankel

n = 0 # tipo 1

# x terá de ser igual a |r-rs|*k
# k já está definido

def hankel(n,k,x,y,ys):
    r_rs = np.sqrt(x**2+(y-ys)**2)
    r_rs_k = r_rs*k
    return scipy.special.hankel1(n,r_rs_k)

ys = np.linspace(-15,15,50)
x = 10
y = 0
n = 0
# Função Integranda

def integrand(ys):
    return hankel(n,k,x,y,ys)*U(x,y)

absoluto = np.sqrt(np.real(integrand(ys))**2+np.imag(integrand(ys))**2)

plt.plot(ys,np.real(integrand(ys)),label="Real")
plt.plot(ys,np.imag(integrand(ys)),label="Imag")
plt.plot(ys,absoluto,label="Magnitude")
plt.legend()
plt.title("Função Integrand")
plt.xlabel("y")
plt.ylabel("f(y)")
plt.show()
```

2. Escreva uma rotina que implemente o método de Simpson para calcular este integral. Represente módulo do erro como função do número de intervalos N até atingir a lei assimptótica. Comente os resultados.

##### O método de Simpson requer, para atingir a lei assimptótica, que o número de pontos seja superior a N = 500, demasiado elevado, pesado e demorado para execução.

```python
# Método de Simpson

def simpson(f,a,b,N):
    h = (b-a)/2
    soma_impar = 0
    soma_par = 0
    for i in range(1,N,2):
        soma_impar += f(a+i*h)
    for i in range(2,N-1,2):
        soma_par += f(a+i*h)
    return h*(f(a)+4*soma_impar+2*soma_par+f(b))*(1/3)

valor_real = 0.69247699620405105797219086532549115 - 0.72131043233017755094937790855952844*1j
N = [1,2,4,8,16,32,64,128,256,1024,2048]

erro_simpson = np.zeros(len(N))
a = -15
b = 15

for i in range(len(N)):
    erro_simpson[i] = np.abs(valor_real-simpson(integrand,a,b,N[i]))

plt.plot(N,erro_simpson)
plt.title("Erro Simpson")
plt.xlabel("N")
plt.ylabel("Erro")
plt.show()
```

3. Escreva uma rotina que implemente método de Romberg. Represente o valor absoluto do erro como função do número de intervalos para cada coluna (diferentes ordens de extrapolação). Comente o que observa. 

##### O erro de Romberg rapidamente converge para o valor de erro independente de N. Também se nota que apartir da 9ª coluna de extrapolação a variação é cada vez menor e menos significativa.

```python
# Romberg

def romberg(f,a,b,tol,iterations):
    R = np.zeros((iterations,iterations))
    h = b-a
    R[0,0] = (b-a)*(f(a)+f(b))/2
    for i in range(1,iterations):
        h = h/2
        R[i,0] = R[i-1,0]/2
        for j in range(1,i+1):
            R[i,j] = R[i,j-1]+(R[i-1,j-1]-R[i-1,j])*(h**j)
            if abs(R[i,j]-R[i,j-1]) < tol:
                return R[i,j], R
            
erro_romberg = np.zeros(len(N))
tol = 1e-10
coluna = np.zeros(len(N))

for i in range(4,len(N)):
    erro_romberg[i] = np.abs(valor_real-romberg(integrand,a,b,tol,N[i])[0])

plt.plot(N,erro_romberg)
plt.title("Erro Romberg")
plt.xlabel("N")
plt.ylabel("Erro")
plt.show()

matriz_romberg = [romberg(integrand, a, b, tol, n) for n in N]
num_plots = len(N) - 4  # Ajusta o número de subplots
fig, axs = plt.subplots(num_plots // 2 + num_plots % 2, 2, figsize=(10, 8))

for i in range(4, len(N)):
    coluna = matriz_romberg[i][1][:i+1, i]
    ax = axs[(i-4) // 2, (i-4) % 2]  # Determina a posição do subplot
    ax.plot(range(1, i+2), coluna)
    ax.set_title(f'Coluna {i+1}')

plt.tight_layout()
plt.show()
```

4. Escreva uma rotina que implemente o integral recorrendo à quadratura com polinómios de Hermite. Comente os resultados.

##### O erro de Hermite aparenta não poder ser calculado apartir de um dado N, compromentendo assim a sua utilização.

```python
def hermite(f, n):
    x, w = hermgauss(n)
    integral = np.sum(w * f(x) * np.exp(x**2))  # Usando a fórmula de quadratura de Gauss-Hermite
    return integral

erro_hermite = np.zeros(len(N))

for i in range(len(N)):
    erro_hermite[i] = np.abs(valor_real-hermite(integrand,N[i]))

plt.plot(N, np.imag(erro_hermite), label='imag')
plt.plot(N, np.real(erro_hermite), label='real')
#plt.plot(N, np.abs(erro_hermite), label='abs')
plt.legend()
plt.title('Erro de Hermite')
plt.xlabel('N')
plt.ylabel('Erro')
plt.show()
```

5. Escreva uma rotina que implemente o integral recorrendo à quadratura de Chebyshev. Comente os resultados.

##### Esta quadratura varia significativamente para valores de N pequenos, mas, apartir de unm valor de N bastante baixo, estabiliza imediatamente.

```python
def cheb(func, n):
    x, w = chebgauss(n)
    integral = np.sum(w * func(x))
    return integral

erro_cheb = np.zeros(len(N))
U_s = cheb(integrand,124)

for i in range(len(N)):
    erro_cheb[i] = np.abs(valor_real-cheb(integrand,N[i]))

plt.plot(N, erro_cheb)
plt.title('Erro de Chebyshev')
plt.xlabel('N')
plt.ylabel('Erro')
plt.show()
```
