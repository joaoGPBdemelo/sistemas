```python
L = 50
i = np.arange(1,L + 1)
j = np.arange(1,L + 1)
k = np.arange(1,L + 1)
i_s = i**2
j_s = j**2
k_s = k**2
total_combinacoes = 3 * np.math.comb(L, 3)
soma = np.zeros(total_combinacoes + 1)
p = 0
for t in  i:
    for m in j:
        for b in k:
            soma[p] += t**2 + m**2 + b**2
            if ((t + m+ b) % 2 != 0):
                soma[p] *= -1
            p+=1


raiz = np.sqrt(soma)

inverso = 1/raiz
M = np.sum(inverso)


print(M)
```

```python
L = 100
i = np.arange(-L,L+1)
j = np.arange(-L,L+1)
k = np.arange(-L,L+1)
soma = 0

for y in i:
    for x in j:
        for z in k:
            if((x + y + z) == 0):
                    continue
            if ((x + y +z) % 2 == 0):
                soma += 1/(np.sqrt(x**2 + y**2 + z**2))
            else:
                soma -= 1/(np.sqrt(x**2 + y**2 + z**2))
print(soma)
```

```python
# coefbinomial a) 
def binomial(n,k):
  
    if (k == 0):
        return 1
    ffn = 1
    ffk = 1
    ffn_k = 1
    for i in range(1, n + 1):
        ffn *= i
    for j in range (1,k + 1):
        ffk *= j
    for z in range (1,n - k + 1):
        ffn_k *= z
    return(ffn//(ffk*ffn_k))
print(binomial(10,3))
```

```python
# b) 
l = 20
for n in range (1,l+1):
    k = 0
    for k in range (n + 1):
        print(binomial(n,k))
    print("\n")
```

```python
# b) 
l = 20
print(1,"\n")
for n in range (1,l+1):
    cn = 1
    for k in range (n+1):
        if (k == 0):
            print(binomial(n,0),end = ' ')
        else:
            cn = cn*1*(n - (k-1))//k 
            print(cn,end = ' ')
    print("\n")
#lembrar que na recursividade tenho que colocar o outro vezres o anterior
#não basta colocar o primeiro k vezes o outro
```

```python
#c) 

prob  = binomial(100,60)/2**100
print(prob)
```

```python
# calculando primos
M = 104729
primos = []
k = 2
for i in range (2,M + 1):
    k = 2
    while k < np.sqrt(i):
        if (i % k == 0):
            break
        k += 1
    if (k > np.sqrt(i)): #lembrar que aqui ele ja adicionou o 1
        primos.append(i)
print(primos)
```

```python
import numpy as np

L = 100
i, j, k = np.arange(-L, L+1), np.arange(-L, L+1), np.arange(-L, L+1)

x, y, z = np.meshgrid(i, j, k)
xyz_sum = x**2 + y**2 + z**2
nonzero_indices = np.nonzero(xyz_sum)

soma = np.sum(np.where(xyz_sum[nonzero_indices] % 2 == 0,
                       1/np.sqrt(x[nonzero_indices]**2 + y[nonzero_indices]**2 + z[nonzero_indices]**2),
                       -1/np.sqrt(x[nonzero_indices]**2 + y[nonzero_indices]**2 + z[nonzero_indices]**2)))

soma
```

```python
L = 100
i,j,k = np.arange( -L, L+1 ), np.arange( -L, L+1 ), np.arange( -L, L+1 )
# cira 3 array com elemento 1 a 1 de -L ate L
x,y,z = np.meshgrid(i,j,k)
# cria array 3 array de dim iguais de modo que quando somalos terei uma array 3D com todas as combinações de termos

comb_xyz = x**2 + y**2 + z**2
# quero todas as combinaçoes possíveis dos quadrados

indices_n_zero = np.nonzero(comb_xyz)
#o valor de todos os indices os quais minha matriz 3d não vale zero
# e os armazena em tuplas para quando fizer minhas combinações poder botar os indices

soma = np.sum(np.where(comb_xyz[indices_n_zero] % 2 == 0,
               + 1/np.sqrt((x[indices_n_zero]**2 + y[indices_n_zero]**2 + z[indices_n_zero]**2)),
              - 1/np.sqrt((x[indices_n_zero]**2 + y[indices_n_zero]**2 + z[indices_n_zero]**2))))
# armazena numa variável a soma dos valores de dentro, com uma condicional, quando minha matriz 3D fizer
# uma soma par soma-se e impar se subtrai.

soma
```
