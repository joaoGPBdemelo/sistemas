## Pergunta 1
**Princípio de Huygens** - Christiaan Huygens (1629-1685) foi um físico, matemático e engenheiro Holandês que propôs um método de representação de frentes de onda, onde cada ponto de uma frente de onda se comporta como uma nova fonte de ondas elementares, que se propagam para além da região já atingida pela onda original e com a mesma frequência. Este ideia é conhecida por princípio de Huygens. Apesar de não ser exacta e ter vindo a sofrer sucessivas alterações (Fresnel, Kirchhoff), com o passar dos anos, ainda é ensinada pois permite introduzir alguns dos conceitos fundamentais da interferência de ondas.

Vamos considerar que um feixe laser é emitido de uma fonte colocada num ponto X com uma amplitude dada por um perfil gaussiano, 
\begin{equation}
U(x,y)=U_{0}e^{ikx}e^{-\frac{y^{2}}{2\sigma^{2}}},
\end{equation}
onde $k$ é número de onda em $xx'$ ($k=2\pi/\lambda$) e $\sigma$ é uma medida da largura do feixe. Neste problema iremos usar $\sigma=1$. Para aplicarmos o princípio de Huygens, vamos considerar una superfície $\Sigma$ que corta o feixe do laser. Podemos imaginar que cada ponto da superfície $(x_{s},y_{s})$, emite uma radiação circular (2D). Em qualquer ponto do espaço, a amplitude medida será dada pela sobreposição de todas as amplitudes das onduletas geradas pela superfície $\Sigma$ (a 2D é uma linha). A figura seguinte tem um esquema que representa esta construção.



Desta forma a amplitude secundária num ponto $\boldsymbol{r}=(x,y)$, é dada por 
\begin{equation}
U_{s}(x,y)=\frac{k}{2}\int_{\Sigma}dr_{s}U(x_{s},y_{s})H_{0}^{(1)}(\left|\boldsymbol{r}-\boldsymbol{r}_{s}\right|k)
\end{equation}
onde $H_{0}^{(1)}(x)$ é uma função de Hankel do primeiro tipo que pode ser chamada a partir do [https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.hankel1.html||scipy.special.hankel1] que resultou a integração do príncipio de Huygens em z para obter a teoria 2D. Considere $k=512$, $\sigma=1$ e $U_{0}=1$. Pretendemos calcular $U_{s}(10,0)$ considerando que a superfície $\Sigma$ é a linha infinita que passa em x=0. O integral é numérico dá (exacto a 25 casas decimais), 
\begin{equation}
U_{s}(10,0)=0.69247699620405105797219086532549115-0.72131043233017755094937790855952844i.
\end{equation}
1. Represente a função integranda no intervalo de $y_{s}$ de $[-15,15]$. Explique como sugere lidar com o limite de integração. 

O domínio de integração é $y_s \in [-\infty, + \infty]$, contudo como o feixe tem uma modulação gaussiana, vemos que amplitude da função integranda, decresce muito rapidamente para zero. Em $y_s=15$ o valor da função já é da ordem de $10^{-50}$ muito abaixo do que os nossos métodos numericos são capazes de ter em precisão. Por este motivo, com funções tão rapidamente decrescentes, faz sentido limitar o domínio a $[-\Delta, \Delta]$ controlado por uma parâmetro de *cutoff* $\Delta$ e deprezar as contribuições exteriores ao intervalo.   

```python
def U(x,y, k):
    return np.exp(1j* k * x - 0.5*y**2 )

def Ui(x,y,xs,ys,k):
    r = np.sqrt((x - xs)**2 + (y - ys)**2 )
    return sp.hankel1(0, k * r) * k/2 * U(xs,ys,k)


ys = np.linspace(-15, 15, 16384)
uus = Ui(10,0,0, ys,512)

fig, axs = plt.subplots(3, 2,figsize=(15,10))
axs[0,0].plot(ys, np.real(uus) )

axs[1,0].plot(ys, np.imag(uus) )
axs[2,0].plot(ys, np.abs(uus) )

axs[0,1].plot(ys, np.real(uus) )
axs[1,1].plot(ys, np.imag(uus) )
axs[2,1].plot(ys, np.abs(uus) )

axs[2,0].set_xlabel(r"$y_s$", fontsize=14)
axs[2,1].set_xlabel(r"$y_s$", fontsize=14)

fig.tight_layout()
ylabel = ["Re(f)", "Im(f)", "Abs(f)"]
for i in range(3):
    axs[i,1].set_xlim(-4,4)
    axs[i%2,0].set_xticks([],[])
    axs[i%2,1].set_xticks([],[])
    axs[i,1].set_yticks([],[])
    axs[i,0].set_ylabel(ylabel[i], fontsize=14)
```

2. Escreva uma rotina que implemente o método de Simpson para calcular este integral. Represente módulo do erro como função do número de intervalos N até atingir a lei assimptótica. Comente os resultados.

Implementamos o método método de Simpson tirando partido das capacidades de vectorização do numpy numa função onde indicamos o intervalo de integração e o número de intervalos. Tirando partido de sabermos o valor "exacto" do integral podemos calcular o erro,
\begin{equation}
erro(N) = \left| U_s(\Delta = 15, N) - exact(\Delta) \right|
\end{equation}
A teoria diz que assimptoticamente o erro deverá decrescer com a potência $h^4$. Na figura a) representamos o erro junto com uma linha com a potência de $h^4$ para comparar. Claramente, o erro não decresce com uma lei de potência (em escala log-log é uma linha recta). 

**Porque motivo não é possível observar a lei de potência?**

Vimos na teoria que o integral de uma função é dado pelo somatório com a regra de Simpson mais uma série infinita de potências pares iguais ou maiores que quatro com coeficientes que dependem do integral exacto de derivadas da função integranda mais o erro de truncatura.
Em particular o coeficiente do termo em $h^4$ é proporcional a
\begin{equation}
\int_a^b dx f^{(4)}(x) = f^{(3)}(b) - f^{(3)}(a)
\end{equation}

Esta diferença para $\Delta = 15$ é minúscula, $10^{-50}$, muito inferior ao erro de truncatura.
Contudo, esta parcela é dominante para $h$ suficientemente pequeno. Com $h$ finito, as ordens superiores da expansão podem ter uma contribuição superior, uma vez que a a derivada da função integranda é inicialmente crescente com a ordem da derivada.
Por este motivo, para um $h$ fixo e finito é possível que a contribuição da mais importante da série não seja das ordens mais baixas.

É precisamente isso que acontece com esta função. 

Para testar esta hipótese, podemos diminuir o parâmetro $\Delta$ de modo a ter a derivada de terceira ordem superior nos limites do intervalo. 
Na figura b) representamos o erro em função de um valor $\tilde{N}$ de número de intervalos por comprimento do intervalo (note-se que estamos a variar o *cutoff* ). Para $\Delta =  1$ podemos ver uma clara região $h^4$ a emergir a partir de $\tilde{N} = 1024$. Quando aumentamos o valor de $\Delta$, esta região aparece cada vez para valores de $\tilde{N}$ superiores. A região com a lei de potência acaba quando o erro do método é da ordem do erro de truncatura dos números de vírgula flutuante. 
Para $\Delta = 8$ já não é possível observar a região com a lei de potência adequada. Seria necessário, trabalhar com números com cerca de 50 algarismos e um número de intervalos muito suuperior para ser possível observar a lei de potência.

```python
def Simpson(func, a, b, N):
    h = (b - a)/N
    x = np.linspace(a,b, N + 1)
    f = func(x)
    return h/3*( f[0] + 4 * f[1:-1:2].sum() + 2 * f[2:-2:2].sum() + f[-1])

Ns = 2**np.arange(4,20)
erro = np.zeros(len(Ns))

figs, axs = plt.subplots(1,2,figsize=(10,5))

exact = 0.69247699620405105797219086532549115-0.72131043233017755094937790855952844j
for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -15, 15, N)/exact - 1)
axs[0].plot(Ns, erro,'--', label=r'$\Delta = 15$')
axs[0].plot(Ns[:-2], (100/Ns[:-2])**4.0, 'k--', label=r"Guia de potência $1/x^4$")


exact =  0.62897348465964128048495946887753239-0.74477220143949755091160819966307907j


for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -1, 1, N)/exact - 1)
axs[1].plot(16 * Ns, erro, label=r'$\Delta = 1$' )

exact = 0.68733397431669297939716133607256783-0.72692915481201960110833033686360705j
for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -2, 2, N)/exact - 1)
axs[1].plot(8*Ns, erro, label=r'$\Delta = 2$')


exact = 0.69247596984193045107059513683964150-0.72130078398187939729096173104513005j
for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -4, 4, N)/exact - 1)
axs[1].plot(4*Ns, erro, label=r'$\Delta = 4$')

exact = 0.69247699620405115639579459135137135-0.72131043233017737719531549257806131j
for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -8, 8, N)/exact - 1)
axs[1].plot(2*Ns, erro, label=r'$\Delta = 8$')

exact = 0.69247699620405105797219086532549115-0.72131043233017755094937790855952844j
for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -16, 16, N)/exact - 1)
axs[1].plot(Ns, erro,'--', label=r'$\Delta = 16$')
axs[1].plot(Ns[:-2], (100/Ns[:-2])**4.0, 'k--', label=r"Guia de potência $1/x^4$")

title = ["a)", "b)"]
for i in range(2):
    axs[i].set_xscale('log')
    axs[i].set_yscale('log')
    axs[i].set_title(title[i])

axs[0].set_xlabel("N")
axs[1].set_xlabel(r"$\tilde{N}$")

plt.legend()
```

3. Escreva uma rotina que implemente método de Romberg. Represente o valor absoluto do erro como função do número de intervalos para cada coluna (diferentes ordens de extrapolação). Comente o que observa. 

Vimos na pergunta anterior que devido às propriedades da função integranda o erro é dominado pelas ordens superiores. O método de Romberg é baseado em eliminar as ordens mais baixas através da avaliação do integral numérico em grelhas diferentes. Por este motivo não esperamos nenhuma melhoria com o método de Romberg para $\Delta = 15$.

Contudo, só para efeitos de ver o método a funcionar podemos analisar o comportamento para $\Delta = 1$ porque sabemos que o erro será dominado pelas ordens mais baixas.
É precismente isso que vemos representado. Na figura com $\Delta = 15$ não vemos nenhuma melhoria dos resultados com o aumento da ordem das colunas do Romberg. Pelo contrário, claramente a convergência até se torna mais lenta.

Já no gráfico com $\Delta = 1$ vemos a convergência esperada em cada coluna. O desvio escala em cada coluna com uma potência diferente em $h$. Na coluna correspondente ao método de ordem mais elevada o erro de truncatura é atingido com menos intervalos revelando o aumento de eficiência.

As linhas a tracejado são apenas guias de visuais com as leis de potência adequadas. 


```python
def Trapezio(func, a, b, N):
    h = (b - a)/N
    x = np.linspace(a,b, N + 1)
    f = func(x)
    return h/2*( f[0] + 2 * f[1:-1].sum()  + f[-1])

def romberg(func, a, b, eps):
    table = np.zeros([2, 2], dtype=np.complex128)
    order = 1
    
    table[order - 1,0] = Trapezio(func, a, b, 2**(4 + order + 0))
    table[order,0] = Trapezio(func, a, b, 2**(4 + order + 1))

    erro = ( table[1,0] - table[0,0] )/( 2**(2 * 1) - 1 )
    table[1,1] = table[1,0] + erro
    
    while(abs(erro) > eps ):
        order += 1
        tmp = np.copy(table)
        table = np.zeros([order + 1, order + 1], dtype=np.complex128)
        table[:order, :order] = tmp[:,:]
        
        table[order, 0] = Trapezio(func, a, b, 2**(4 + order + 1))
        
        for i in range(1, order + 1):
            erro = (table[order,i - 1] - table[order-1,i - 1])/(2**(2 * i) - 1)
            table[order, i] = table[order,i - 1] + erro
            
        erro = abs((table[order,order - 1] - table[order-1,order - 1])/(2**(2 * order) - 1))
    return table

fig, axs = plt.subplots(1,2,figsize=(15,5))
teste = romberg(lambda ys: Ui(10,0,0,ys, 512), -15,15, 1.E-16) 
exact = 0.69247699620405105797219086532549115-0.72131043233017755094937790855952844j

Ns = 2**(4 + np.arange(len(teste[:,0])))
axs[0].plot(Ns[:], np.abs(teste[:,0] - exact), label= "1 col.")
axs[0].plot(Ns[1:], np.abs(teste[1:,1] - exact), label= "2 col.")
axs[0].plot(Ns[2:], np.abs(teste[2:,2] - exact), label= "3 col.")
axs[0].plot(Ns[3:], np.abs(teste[3:,3] - exact), label= "4 col.")
axs[0].plot(Ns[4:], np.abs(teste[4:,4] - exact), label= "5 col.")
axs[0].plot(Ns[5:], np.abs(teste[5:,5] - exact), label= "6 col.")
axs[0].set_title(r'$ \Delta = 15 $')
axs[0].set_yscale('log')
axs[0].set_xscale('log')
axs[0].legend()
axs[0].set_xlabel(r'$N$')


teste = romberg(lambda ys: Ui(10,0,0,ys, 512), -1,1, 1.E-16) 
exact =  0.62897348465964128048495946887753239-0.74477220143949755091160819966307907j


Ns = 2**(4 + np.arange(len(teste[:,0])))
axs[1].plot(Ns[:], np.abs(teste[:,0] - exact), label = "1 col.")
axs[1].plot(Ns[:], (3/Ns[:])**2, color='blue', linestyle='dashed', label = r"$C/h^2$" )
axs[1].plot(Ns[1:], np.abs(teste[1:,1] - exact), label = "2 col." )
axs[1].plot(Ns[1:], (8/Ns[1:])**4, color='orange', linestyle='dashed', label = r"$C/h^4$" )
axs[1].plot(Ns[2:], np.abs(teste[2:,2] - exact), label = "3 col.")
axs[1].plot(Ns[2:], (13/Ns[2:])**6, color='green', linestyle='dashed', label = r"$C/h^6$")
axs[1].plot(Ns[3:], np.abs(teste[3:,3] - exact), label = "4 col.")
axs[1].plot(Ns[3:], (19/Ns[3:])**8, color='red', linestyle='dashed', label = r"$C/h^8$")
axs[1].plot(Ns[4:], np.abs(teste[4:,4] - exact), label = "5 col.")
axs[1].plot(Ns[4:], (29/Ns[4:])**10, color='purple', linestyle='dashed', label = r"$C/h^{10}$")

axs[1].set_yscale('log')
axs[1].set_xscale('log')
axs[1].set_title(r'$ \Delta = 1 $')
axs[1].set_xlabel(r'$N$')
axs[1].legend()
```

4. Escreva uma rotina que implemente o integral recorrendo à quadratura com polinómios de Hermite. Comente os resultados.

Utilizamos a rotina do scipy para obter quer os pesos quer os zeros dos polinómios de Hermite. Apesar da função integranda ter uma gaussiana que pode ser associada ao peso dos polinómios de Hermite, não há uma clara vantagem no uso destes polinómios. A função está concentrada em torno de $y_s = 0$ e é fortemente oscilante. Como a região de integração com os polinómios de Hermite é infinita, com o aumento da ordem dos polinómios, existem cada vez mais zeros fora da região de interesse. Isto reflecte-se num não aproveitamento desses pontos.

No gráfico a) observamos o erro com o número observamos que o erro diminui com o aumento de pontos mas não é vantajoso face ao método de Simpson. 
No gráfico b) podemos observar que a fracção de pontos entre $[-15, 15]$ está a diminuir com o aumento da ordem/pontos da quadratura.


```python
def UiH(x,y,xs,ys,k):
    r = np.sqrt((x - xs)**2 + (y - ys)**2 )
    return sp.hankel1(0, k * r) * k/2 * np.exp(1j*xs*k)

exact =  0.69247699620405105797219086532549115 - 0.72131043233017755094937790855952844j
Ns = 2**np.arange(4,19)
erro = np.zeros(len(Ns))
npontos = np.zeros(len(Ns))
for i, N in enumerate(Ns):
    ys, w = sp.roots_hermitenorm(N)
    npontos[i] = len(np.where(np.abs(ys) < 15 )[0])
    erro[i] = np.abs((UiH(10,0,0,ys, 512 ) @ w) /exact - 1)
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].plot(Ns, erro,'o-', label="Quad. Hermite")

for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -15, 15, N)/exact - 1)
axs[0].plot(Ns, erro,'o-', label="Simpson Method")

axs[1].plot(Ns, npontos/Ns,'o-', label="Fracção de pontos em [-15,15]")

title = ["a)", "b)"]
for i in range(2):
    axs[i].set_yscale('log')
    axs[i].set_xscale('log')
    axs[i].set_title(title[i])
    axs[i].set_xlabel(r'$N$')
    axs[i].legend()
```

5. Escreva uma rotina que implemente o integral recorrendo à quadratura de Chebyshev. Comente os resultados.

Usando a quadratura de Chebyshev podemos aproximar o integral por,
\begin{equation}
\int_{-1}^{1} dx \, f(x) \frac{1}{\sqrt{1 - x^2}} \approx \sum_{i = 0}^{N-1} w_i f(x_i)
\end{equation}
Onde os $w_i$ são os pesos e $x_i$ são os zeros do polinómio de ordem $N$
Neste caso nem o domínio de integração é o intervalo $]-1,1[$ nem a função integranda tem o peso explicitamente. 
Para fazer a conversão basta redefinir, 
\begin{align}
\int_{a}^{b} dx g(x) & = \frac{b - a}{2} \int_{-1}^{1} dx' g\left( \frac{a + b}{2} + \frac{b - a}{2} x' \right) \\
& = \frac{b - a}{2} \int_{-1}^{1} \frac{dx'}{\sqrt{1 - {x'}^2 }} g\left( \frac{a + b}{2} + \frac{b - a}{2} x' \right) \sqrt{1 - {x'}^2} \\
& \approx \sum_{i = 0}^{N-1} w_i g\left( \frac{a + b}{2} + \frac{b - a}{2} {x_i'} \right) \sqrt{1 - {x_i'}^2}
\end{align}

Verificamos que em condições semelhantes este método converge um pouco melhor.

```python
def integral_Chebyshev(x,y, a, b, N):
    ys, w = sp.roots_chebyt(N)
    yt = (a + b)/2 + (b - a)/2 * ys
    f = Ui(x,y, 0, yt, 512) * np.sqrt(1 - ys**2) * (b - a)/2    
    return f @ w


exact =  0.69247699620405105797219086532549115 - 0.72131043233017755094937790855952844j
Ns = 2**np.arange(4,19)
erro = np.zeros(len(Ns))

plt.figure(figsize=[10,5])
for i, N in enumerate(Ns):
    erro[i] = np.abs(integral_Chebyshev(10,0, -15, 15, N)/exact - 1)
plt.plot(Ns, erro,'o-',label='Quad. Chebyshev')

for i, N in enumerate(Ns):
    erro[i] = np.abs(Simpson(lambda ys: Ui(10,0,0,ys, 512), -15, 15, N)/exact - 1)
plt.plot(Ns, erro,'o-', label="Simpson Method")
plt.xlabel(r'$N$',fontsize=14)
plt.legend()
plt.yscale('log')
plt.xscale('log')
```

**Principio de Babinet** - O enunciado do princípio de Babibet é

O padrão de difração observado quando a luz incide sobre uma abertura de qualquer forma é o mesmo obtido quando a luz incide sobre um objeto que é o complemento da abertura.

Para verificar o princípio de Babinet iremos comparar as amplitudes com um obstáculo e com uma fenda e comparar com perfil livre.

Calculamos o campo em cada ponto com $x=x_p$ somando as contibuições de $[-15, -0.5]$ com $[0.5,15]$ ou seja no obstáculo não há emissores de secundárias. 

```python
ya = np.linspace(-4,4, 1024 + 1)
resultO = 1j * ya
for i, ya1 in enumerate(ya):
    resultO[i] = integral_Chebyshev(10,ya1, -15, -0.5, 4096) + integral_Chebyshev(10,ya1, 0.5, 15, 4096)
```

O resultado do campo gerado pela fenda é dado pela integração entre $[-0.5,0.5]$ onde estão os pontos emissores. 

```python
ya = np.linspace(-4,4, 1024 + 1)
resultF = 1j * ya

for i, ya1 in enumerate(ya):
    resultF[i] = integral_Chebyshev(10,ya1, -0.5, 0.5, 4096)
```

## Pergunta 3
**Lei de Snell** - Considere a seguinte montagem,



Agora laser tem um feixe paralelo à direcção x' e uma variação da amplitude com y', U(x',y'). O feixe, incide numa superfície delimitada por x=0, e propaga-se com um k'=1.2\,k.

1. Represente a amplitude U(x',y') para x=0 , ou seja, no plano da interface para um ângulo \theta=\pi/6.


```python
def Uq(x,y, k, theta):
    xl = - y *np.sin(theta)
    yl = y * np.cos(theta)
    return np.exp(1j* k * xl - 0.5*yl**2 )

def Uiq(x,y,xs,ys,k,theta):
    r = np.sqrt((x - xs)**2 + (y - ys)**2 )
    return sp.hankel1(0, 1.2 * k * r) * 1.2 * k/2 * Uq(xs,ys,k,theta)


y = np.linspace(-6,6, 16384)
angulos = [r'\pi/4', r'\pi/6', "0" ]
fig, axs = plt.subplots(1,2,figsize=(15,5))
Thetas = np.array([np.pi/4, np.pi/6, 0])
for i,theta in enumerate(Thetas):
    axs[0].plot(y, np.real(Uq(0,y, 512,theta)), label=r"$\theta = "+ angulos[i] + "$" )
    axs[1].plot(y, np.real(Uq(0, y , 512,theta)), label=r"$\theta = "+ angulos[i] + "$" )
    axs[1].set_xlim([-2,0])
axs[0].legend()
axs[1].legend()
```

A relação entre as coordenadas em ambos referenciais é dada por,
\begin{align}
x' & = x \cos (\theta) -y \sin(\theta) \\
y' & = x \sin (\theta) + y \cos(\etheta)
\end{align}
e
\begin{align}
x & = x' \cos (\theta) + y' \sin(\theta) \\
y & = - x' \sin (\theta) + y' \cos(\theta)
\end{align}

Usando estas coordenadas podemos escrever o campo gerado em $(x = 0, y)$ para diferentes valores de $\theta$. Podemos ver que para $\theta = 0$ o campo primário é uma função suave. Contudo, com o ângulos finitos aparece uma oscilação associada à componente do vector de onda na direção da linha.  

2. Represente o módulo da amplitude no alvo a x_{p}=10. Consegue extrair o ângulo \theta'?

```python
def integral_ChebyshevB(x,y, N, theta):
    a = -32
    b =  32 
    ys, w = sp.roots_chebyt(N)
    yt = (a + b)/2 + (b - a)/2 * ys
    f = Uiq(x, y, 0, yt, 512, theta) * np.sqrt(1 - ys**2) * (b - a)/2    
    return f @ w

yps = np.linspace(-15,5,512)
U = 0.0 * yps
theta =  np.pi/6
for i,yp in enumerate(yps):
    U[i] = np.abs(integral_ChebyshevB(10,yp, 16384,theta))
plt.vlines(-np.tan(np.pi/6)*10, ymin=0, ymax = 1.2, colors='k',linestyles='dashed')
plt.plot(yps, U)
plt.xlabel(r"$y$", fontsize=14)
plt.ylabel(r"$abs(U)$", fontsize=14)
print("O máximo é dado por: ", np.arctan(- yps[np.argmax(U)]/10), " em vez de ", np.pi/6) 
```

Calculamos o campo na superfície $x = 0$ e propagamos usando a construção de Huygens para $x>0$. Nessa região devido ao índice de refracção ser $n = 1.2 \ne 1.0$. Analisando o módulo do campo em $x = 10$, vemos que máximo está deslocado em relação ao ângulo $\theta = \pi/3$ que representamos a tracejado. O máximo corresponde a valor de ângulo inferior tal como era esperado pela lei de refracção.

3. Repita o procedimento anterior para vários valores de $0<\theta<\pi/2$. Represente \sin(\theta') como função de $\sin(\theta)$ e compare com a previsão da lei de Snell.  

```python
Thetas = np.linspace(0.05, np.pi/2 - 0.05, 8)
Thetasl = 0.0 *Thetas

yps = np.linspace(-35,5,256)
U = 0.0 * yps

for j, theta in enumerate(Thetas): 
    for i,yp in enumerate(yps):
        U[i] = np.abs(integral_ChebyshevB(10,yp, 16384,theta))
    Thetasl[j] = np.arctan(- np.sum(yps * U)/np.sum(U) /10 )

plt.plot(np.sin(Thetas),  np.sin(Thetasl), 'o', label=r"$f_1(\sin(\theta_1))$")
plt.plot(np.sin(Thetas),  np.sin(Thetas)/1.2, '--', label=r"$f_2(\sin(\theta_1))$" )
plt.xlabel(r"$\sin(\theta_1)$")
plt.legend()
```

A lei de Snell relaciona os ângulos de propagação antes e depois de um feixe incidir sobre uma superfície que altera índice de reflexão. 
\begin{equation}
n_1 \sin(\theta_1) = n_2 \sin(\theta_2)
\end{equation}

Para verificar esta lei vamos calcular o valor médio da coordenada $y$ usando um peso relativo proporcional ao módulo do campo. 
Sabendo a distância entre os planos podemos calcular o ângulo $\theta_2$. 
Na figura representamos as funções:
- $f_1(\sin(\theta_1)) = \sin(\theta_2)$
- $f_2(\sin(\theta_1)) = \frac{\sin(\theta_1)}{n_2} $

Para cada valor de $\theta_1$ medimos $\theta_2$ e comparamos com a previsão teórica. Verificamos que a lei de Snell é verificada.
