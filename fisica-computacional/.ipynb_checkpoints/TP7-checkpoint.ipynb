### Exerc√≠cio 1

```python
def rk4(a,b,N,x_0,f):
    h = (b-a)/N
    tpoints = np.linspace(a + 1e-16,b,N) 
    xpoints = np.zeros(N)
    x = x_0 
    for i,t in enumerate(tpoints):
        xpoints[i] = x
        k1 = h*f(x,t)
        k2 = h*f(x+0.5*k1,t+0.5*h)
        k3 = h*f(x+0.5*k2,t+0.5*h)
        k4 = h*f(x+k3,t+h)
        x += (k1+2*k2+2*k3+k4)/6
    return xpoints

def onda_quadrada(x, amplitude, periodo):
    onda = amplitude * np.sign(np.sin(2*np.pi*x/periodo))
    return onda
```

```python
N = 10000
t = np.linspace(1e-16,10,N)
def f1(vout,t):
    vin = onda_quadrada(t,1,1) 
    return(1/0.01)*(vin - vout)

def f2(vout,t):
    vin = onda_quadrada(t,1,1) 
    return(1/0.1)*(vin - vout)

def f3(vout,t):
    vin = onda_quadrada(t,1,1) 
    return (vin - vout)

vout_0 = 0
vout1 = rk4(0,10,N,vout_0,f1)
vout2 = rk4(0,10,N,vout_0,f2)
vout3 = rk4(0,10,N,vout_0,f3)
plt.plot(t,vout1)
```

```python
plt.plot(t,vout2)
```

```python
plt.plot(t,vout3)
```

### Exerc√≠cios 3

```python
def rk4_3d(a,b,N,r_0,f):
    h = (b-a)/N
    tpoints = np.linspace(a + 1e-16,b,N) 
    xpoints = np.zeros(N)
    ypoints = np.zeros(N)
    zpoints = np.zeros(N)
    r = r_0
    for i,t in enumerate(tpoints):
        xpoints[i] = r[0]
        ypoints[i] = r[1]
        zpoints[i] = r[2]
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1,t+0.5*h)
        k3 = h*f(r+0.5*k2,t+0.5*h)
        k4 = h*f(r+k3,t+h)
        r += (k1+2*k2+2*k3+k4)/6
    return [xpoints,ypoints,zpoints]
```

```python
sigma = 10
erre = 28
b = 8/3
def f(r,t):
    fx = sigma*(r[1] - r[0])
    fy = erre*r[0] - r[1] - r[2]*r[0] 
    fz = r[0]*r[1] - b*r[2]
    return np.array([fx,fy,fz])

a = 0
b_f = 50
N = 10000
t = np.linspace(0,50,N)
r_0 = (0,1,0)

r = rk4_3d(a,b_f,N,r_0,f)
x = r[0]
y = r[1]
z = r[2]
plt.plot(t,y)

```

```python
plt.plot(x,z)
```

### Exerc√≠io 2

```python
def rk4_2d(a,b,N,r_0,f):
    h = (b-a)/N
    tpoints = np.linspace(a + 1e-16,b,N) 
    xpoints = np.zeros(N)
    ypoints = np.zeros(N)
    r = r_0
    for i,t in enumerate(tpoints):
        xpoints[i] = r[0]
        ypoints[i] = r[1]
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1,t+0.5*h)
        k3 = h*f(r+0.5*k2,t+0.5*h)
        k4 = h*f(r+k3,t+h)
        r += (k1+2*k2+2*k3+k4)/6
    return [xpoints,ypoints]
```

```python
alpha = 1
beta = 0.5
gamma = 0.5
delta = 2

def f(r,t):
    fx = alpha*r[0] - beta*r[0]*r[1]
    fy = gamma*r[0]*r[1] - delta*r[1] 
    return np.array([fx,fy])

x = 2
y = 2
r_0 = (x,y)
a = 0
b_f = 30
N = 10000
t = np.linspace(a,b,N)


vec = rk4_2d(a,b,N,r_0,f)
plt.plot(t,vec[0])
```

```python
plt.plot(t,vec[1])
```
