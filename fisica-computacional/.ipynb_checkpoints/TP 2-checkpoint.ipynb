```python
fig, ax = plt.subplots(1,2)

x = data[0:1000,0]# apenas o conteudo da primeira coluna
y = data[0:1000,1]# agora √© o conteudo da segunda coluna

ax.scatter(x , y,marker ='o', color = 'blue', facecolor = 'w')
ax.set(title = "Manchas Solares")

plt.show()
```

```python
# c√°lculo das medias moveis
r = 5
N = 1000
Y = np.zeros(N)

for i in range(-r,r+1):
    som = data[5+i : N+5+i , 1]
    Y += som
    
media_m = (1/(2*r + 1))*Y

fig, ax = plt.subplots()

x = data[5:1005,0]# apenas o conteudo da primeira coluna
y = data[5:1005,1]# agora √© o conteudo da segunda coluna

ax.scatter(x , y,marker ='o', color = 'blue', facecolor = 'w')
ax.set(title = "Manchas Solares")
ax.plot(x,media_m,color = 'red')

plt.show()

```

```python
import numpy as np
from vpython import *
scene = canvas()
#sol
rsol = 1.391
sol = sphere(pos=vector(0,0,0),radius=rsol,color = color.yellow) 
# mecurio
rmer = 0.0244
romer = 3*0.579
vmer = 1
mercurio = sphere(pos=vector(romer,0,0),radius=rmer,color = color.red)
#venus
rv = 0.06052
rov = 3*1.082
vv = 0.391
venus = sphere(pos=vector(rov,0,0),radius=rv,color = color.orange)
#terra
rt = 0.06371
rot = 3*1.496
vt = 0.241
terra = sphere(pos=vector(rot,0,0),radius=rt,color = color.blue)
#marte
rm = 0.03386
rom = 3*2.279
vm = 0.128
marte = sphere(pos=vector(rom,0,0),radius=rm,color = color.orange)
#jupites
rj = 0.69173
roj = 3*7.785
vj = 0.0203
jupiter = sphere(pos=vector(roj,0,0),radius=rj,color = color.orange)
#saturno
rs = 0.57316
ros = 3*14.334
vs = 0.008
saturno = sphere(pos=vector(ros,0,0),radius=rs,color = color.blue)

for t in arange(0,1000*pi,0.1):
    rate(10)# o rate s√≥ muda a velocidade da anima√ß√£o n√£o da minhas esfera
    # mec√∫rio
    x = romer*np.cos(vmer*t)
    y = romer*np.sin(vmer*t)
    mercurio.pos = vector(x,y,0)
    #venus
    x2 = rov*np.cos(vv*t)
    y2 = rov*np.sin(vv*t)
    venus.pos = vector(x2,y2,0)
    #terra
    x3 = rot*np.cos(vt*t)
    y3 = rot*np.sin(vt*t)
    terra.pos = vector(x3,y3,0)
    #marte
    x4 = rom*np.cos(vm*t)
    y4 = rom*np.sin(vm*t)
    marte.pos = vector(x4,y4,0)
    #jupiter
    x5 = roj*np.cos(vj*t)
    y5 = roj*np.sin(vj*t)
    jupiter.pos = vector(x5,y5,0)
    #saturno
    x6 = ros*np.cos(vs*t)
    y6 = ros*np.sin(vs*t)
    
    saturno.pos = vector(x6,y6,0)

```

*Exerc√≠cio 7*

```python
N = 1000
x = np.linspace(-2,2,N)
y = np.linspace(-2,2,N)
r,i = np.meshgrid (x,y)
# vou criar uma grade com cada ponto no meu gr√°fico
c = r + 1j*i

# definir o numero deintera√ß√µes
numi = 100

zx = np.zeros_like(c)
zy = np.zeros_like(c)


for _ in range(numi):
    zx,zy  = np.clip(zx**2 - zy**2 + r,-1e10,1e10) ,np.clip( 2*zx*zy + i,-1e10,1e10)

mandelbrot_set = np.sqrt(zx**2 + zy**2) < 2
plt.imshow(mandelbrot_set, extent=(x.min(), x.max(), y.min(), y.max()), cmap='gray', origin='lower')
plt.colorbar()
plt.title('Conjunto de Mandelbrot')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imagin√°ria')
plt.show()
```
