$$

$$

com a condição de anulamento nas fronteiras $\psi(0)=\psi(L)=0$ devido ao potencial infinito.

Uma base de funções que respeita esta condição fronteira é $\phi_n(x)=\sin(\frac{\pi n}{L}x)$ com $n\in\mathbf{N}$. Estas funções são ortogonais entre si,
$$
\frac{2}{L}\int_0^Ldx\sin(\frac{\pi n}{L}x)\sin(\frac{\pi m}{L}x)=\delta_{nm}.
$$

Escrevendo $\psi(x)$ como combinação linear desta base,
$$
\psi(x)=\sum_{n=1}^{+\infty}\psi_n\sin(\frac{\pi n}{L}x)
$$

podemos transformar a equação dos estados estacionários em
$$
H_{mn}\psi_n=E\psi_m
$$

onde o Hamiltoniano é dado pelos elementos de matriz,
$$
H_{mn}=\frac{2}{L}\int_0^Ldx\sin\left(\frac{\pi m}{L}x\right)\left(-\frac{L^2}{2}\frac{\partial^2}{\partial x^2}+\frac{V(x)}{V_0}\right)\sin\left(\frac{\pi n}{L}x\right)

```python
import numpy as np
def trap (f,N,a,b):
    x = np.linspace(a,b,N+1)
    y = f(x)
    h = np.abs(b-a)/N
    integral = f(a)/2 + f(b)/2 
    for i in range (1,N ):
        integral += f(a + i*h)
    return h * integral

def f1(x, m, n, a, L):
    v0 = np.sqrt(((x/a)-6)**4 + (13/110)*((x/a) - 6)**5)
    func = np.sin(np.pi*m*x/L) * v0 * np.sin(np.pi*n*x/L)
    return func


a = 1
L = 10
a_init = 0
b = L
N = 1000

def QR (A):

    def achar_u (a,q,ordem):
        N = len(a)
        u = np.zeros(N,float)
        som = np.zeros(N,float)
        multi = 0
        for i in range ( ordem + 1):
            multi = np.dot(q[:,i],a)
            som += multi * q[:,i]
        u = a - som
        return u
        
    # Criando as matrizes
    N = len(A)
    R = np.zeros([N,N],float)
    q = np.zeros([N,N],float)
    a_conta = np.zeros([N,N],float)
    a = A.T
    # Fazendo o primeiro caso
    u = A[:,0]

    for i in range(N):

        
        norm_u = np.linalg.norm(u)
        
        q[:,i] = u / norm_u

    
        for j in range (i + 1 , N):
            R[ i , j ] = np.dot(a[j, :] , q[:,i])  
        
        R[i,i] = norm_u
        if i < (N - 1):
            u = achar_u( A[ : ,i + 1 ] , q , i)
        q1 = u / norm_u
    return q,R

def autovalores(A, QR, err=1e-3):# Calcula os autovalores da matriz A  usando o método QR iterativo.
    A = A.copy()
    N = len(A[0])
    V = np.eye(N)
    Q, R = QR(A)
    cont = 0

    while True:
        A = R @ Q
        V = V @ Q
        menor = np.abs(A) < err
        np.fill_diagonal(menor, True)
        if menor.all() or cont >= 1000:
            break
        Q, R = QR(A)
        cont += 1

    return np.diag(A)

# Definindo a base de funções
def phi(n, x, L):
    return np.sin(n * np.pi * x / L)

class Hmn:
    def __init__(self,M):
        
        self.M = M
        self.L = L
        self.a = a
        self.H = np.zeros((M, M), float)
        a_init = 0
        b = L
        N = 1000
        
        for m in range(1, M + 1 ):
 
            for n in range(1 , M + 1):
                if (m-1 == n-1)  :
                    self.H[m-1,n-1] =  trap (lambda x: f1(x,m,n,a,L),N,a_init,b)*L/2 + ((np.pi * n)**2)/L
                else:
                     self.H[m-1,n-1] =  trap (lambda x: f1(x,m,n,a,L),N,a_init,b)*L/2

    def auto (self):   
        return autovalores(self.H,QR)

    
    def calc_psi(self, num_funcs, eigenvectors):
        M = self.M
        L = self.L
        x = np.linspace(0, L, 1000)
        psi = np.zeros((num_funcs, len(x)))
        for i in range(num_funcs):
            for n in range(1, M + 1):
                psi[i,:] += eigenvectors[n - 1, i] * phi(n, x, L)
        return x, psi

             
```

```python
#resolver em x depois em y
def rk4_2d(a,b,N,r_0,f):
    h = (b-a)/N
    tpoints = np.linspace(a + 1e-16,b,N) 
    xpoints = np.zeros(N,complex)
    ypoints = np.zeros(N,complex)
    r = r_0
    for i,t in enumerate(tpoints):
        xpoints[i] = r[0]
        ypoints[i] = r[1]
        k1 = h*f(r,t)
        k2 = h*f(r+0.5*k1,t+0.5*h)
        k3 = h*f(r+0.5*k2,t+0.5*h)
        k4 = h*f(r+k3,t+h)
        r += (k1+2*k2+2*k3+k4)/6
    return [xpoints,ypoints]

# Método Runge-Kutta de quarta ordem
def rk4_4d(a, b, N, r0, f):
    h = (b - a) / N
    t = np.linspace(a, b, N)
    r = np.zeros((4, N))
    r[:, 0] = r0
    
    for i in range(1, N):
        k1 = h * f(r[:, i-1], t[i-1])
        k2 = h * f(r[:, i-1] + 0.5 * k1, t[i-1] + 0.5 * h)
        k3 = h * f(r[:, i-1] + 0.5 * k2, t[i-1] + 0.5 * h)
        k4 = h * f(r[:, i-1] + k3, t[i-1] + h)
        
        r[:, i] = r[:, i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    
    return r
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
const = 20
L = 10  # Em metros
q = 1  # Carga do corpo (em C)
m = 1  # Massa do corpo (em kg)
g = 9.81  # Aceleração gravitacional (em m/s^2)
gamma_m = 0.4  # Razão gamma/m

# Função para calcular o campo elétrico E(r) usando a regra do trapézio
def E_field(r):
    x, y = r
    N = 1000  # Número de pontos para a integração numérica
    x_prime = np.linspace(0, L, N)
    sigma = np.sin(np.pi * x_prime)
    r_prime = np.array([x_prime, np.zeros(N)])
    r_diff_x = x - x_prime
    r_diff_y = y
    r_diff_norm_sq = r_diff_x**2 + r_diff_y**2
    
    integrand_x = sigma * r_diff_x / r_diff_norm_sq
    integrand_y = sigma * r_diff_y / r_diff_norm_sq
    
    Ex = np.trapz(integrand_x, x_prime)
    Ey = np.trapz(integrand_y, x_prime)
    
    return np.array([Ex, Ey]) * const / (2 * np.pi)

# Função RK4 para resolver as equações de movimento
def rk4_step(f, r, t, dt):
    k1 = f(r, t)
    k2 = f(r + 0.5 * dt * k1, t + 0.5 * dt)
    k3 = f(r + 0.5 * dt * k2, t + 0.5 * dt)
    k4 = f(r + dt * k3, t + dt)
    return r + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)

def rk4_4d(f, r0, t):
    dt = t[1] - t[0]
    r = np.zeros((len(t), len(r0)))
    r[0] = r0
    for i in range(1, len(t)):
        r[i] = rk4_step(f, r[i-1], t[i-1], dt)
    return r

# Equações de movimento
def equations_of_motion(r, t):
    rx, vx, ry, vy = r
    v = np.sqrt(vx**2 + vy**2)
    E = E_field([rx, ry])
    fx = vx
    fvx = - gamma_m * v * vx - q * E[0] / m
    fy = vy
    fvy = -g - gamma_m * v * vy - q * E[1] / m
    return np.array([fx, fvx, fy, fvy])

# Condições iniciais
v0_values = [1, 2, 4, 8, 16, 32, 64, 128, 256, 1024]
theta = np.pi / 6
t = np.linspace(0, 5, 1000)  # Intervalo de tempo

# Resolução das equações de movimento para diferentes velocidades iniciais
for v0 in v0_values:
    initial_conditions = [0, v0 * np.cos(theta), 1, v0 * np.sin(theta)]
    solution = rk4_4d(equations_of_motion, initial_conditions, t)
    x, y = solution[:, 0], solution[:, 2]
    # Parar a simulação quando o corpo atingir o solo ou a parede
    mask = (y >= 0) & (x <= L) 
    x = x[mask]
    y = y[mask]




    # Plotar a trajetória
    plt.plot(x, y, label=f'v0 = {v0} m/s')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.legend()
plt.title('Trajetórias para diferentes velocidades iniciais')
plt.grid(True)
plt.show()
```
