```python
def f1(x):
    return np.sin(x)
def f2(x):
    return x * 2
def mandar_fazer(ff,N):
    sum = 0
    for i in range(N):
        sum += ff(i)
    return sum
    
def mandar_fazer2(N): # mais rápido
    sum = 0
    for i in range(N):
        sum += np.sin(i)
    return sum
    
def mandar_fazer3(N):
    sum = 0
    for i in range(N):
        sum += i*2
    return sum

def mandar_fazer4(N):
    return np.sum(np.arange(N,dtype = float)*2)

def mandar_fazer5(N):
    return np.sum(np.sin(np.arange(N,dtype = float)))
                  
%timeit -r 5 -n 2 mandar_fazer(f1,10000000)
# run, loops  
%timeit -r 5 -n 2 mandar_fazer2(10000000)
%timeit -r 5 -n 2 mandar_fazer5(10000000)

    
    
    
    
    
    



    
    
    
    
    
    
    
```
