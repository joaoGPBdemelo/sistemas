2. Considere que a massa central é forçada por aplicação de uma força $\vec{F}=f\boldsymbol{u}_{x}$. Para os valores de $M=5$ e $n=4$, represente no mesmo gráfico a posição de equilíbrio de cada uma das massas ($X_{i}^{eq}$) como função de f. Utilize a eliminação de Gauss Jordan.

\begin{cases}
\frac{d^2 X_1 }{dt^2} & = - (\overline{k}_1 + \overline{k}_2 ) X_1 + \overline{k}_2 X_2 \\
\frac{d^2 X_2 }{dt^2} & = \overline{k}_2 X_1 - (\overline{k}_2 + \overline{k}_3 )X_2 + \overline{k}_3 X_3\\
\dots & \dots \\
\overline{M} \frac{d^2 X_{n+1} }{dt^2} & = + \overline{k}_{n+1} X_n - ( \overline{k}_{n+1} + \overline{k}_{n+2}) X_{n+1} + \overline{k}_{n+2}X_{n+2} + f \\
\dots & \dots\\
\frac{d^2 X_{2n+1} }{dt^2} & = + \overline{k}_{2n+1} X_{2n} - (\overline{k}_{2n+1} + \overline{k}_{2n+2})X_{2n+1}  + \overline{k}_{2n+2}\\
\end{cases}

Existe uma força para a qual a massa central e as massas à direita estarão encostadas à parede direita.

```python
class Molas:
    def __init__(self, n):
        self.n = n
        self.size = 2*n + 1
        self.A = np.zeros([self.size , self.size])
        self.b = np.zeros(2*n + 1)
        self.k = 1 + 0.01*(2 * rd.random_sample(self.size + 1) - 1)
        for i in range(2*n + 1):
            self.A[i,i] = - (self.k[i] + self.k[i + 1])

        for i in range(2*n):
            self.A[i,i+1] = self.k[i+1]
            self.A[i +1,i] = self.k[i+1]
        
        
    def build_matrix_equilibrium(self, f):
        b = np.zeros(self.size)
        b[n] = -f
        b[-1] = - self.k[-1]
        return self.A, b         

    def build_matrix_dynamic(self, M):
        Ad = np.copy(self.A)
        Ad[self.n, :] /= M
        return Ad
   


    


def solveGaussPivotagemTotal(A, b):
    M = len(b)
    linhas = np.arange(M)
    colunas = np.arange(M + 1)
    AA = np.zeros([M, M + 1])
    AA[:M,:M] = A
    AA[:,M] = b
    aa = np.zeros(M * M)

    for p in range(M):
        m = M - p
        for i in range(m):
            aa[i * m:(i + 1) * m] = AA[linhas[p + i], colunas[p:-1]]
        index = np.argmax(np.abs(aa[:m * m]))

        l1 = p + index // m
        c1 = p + index % m
        linhas[p], linhas[l1] = linhas[l1], linhas[p] 
        colunas[p], colunas[c1] = colunas[c1], colunas[p]        
        AA[linhas[p], colunas[:] ] /= AA[linhas[p],colunas[p]]

        for i in range(p):
            AA[linhas[i],colunas[p:]] -= AA[linhas[p],colunas[p:]] * AA[linhas[i],colunas[p]]
            
        for i in range(p + 1, M):
            AA[linhas[i],colunas[p:]] -= AA[linhas[p],colunas[p:]] * AA[linhas[i],colunas[p]]

    return AA[:, -1]

n = 4
Sist = Molas(n)
Mf  = 10 
fs = np.linspace(0,0.2,Mf)
Xeq = np.zeros([Mf, Sist.size])
for i,f in enumerate(fs):
    A, b = Sist.build_matrix_equilibrium(f)
    Xeq[i,:] = solveGaussPivotagemTotal(A,b)

plt.figure(figsize=(12,5))
for i in range(9):
    plt.plot(fs, Xeq[:,i], label=f"$X_"+str(i + 1)+"$")
plt.xlabel(r"$f$",fontsize=16)
plt.ylabel(r"$X_{eq}$",fontsize=16)
plt.legend(bbox_to_anchor=(0.6, 0., 0.5, 0.5) )
```

5. Tente obter os valores próprios desta matriz como função de M usando o método da decomposição QR. Comente os resultados.

O algoritmo de diagonalizar uma matriz através da decomposição QR apenas resulta para matrizes simétricas. No caso de uma matriz não simétrica não existe uma matriz ortogonal que transforme a matriz numa matriz diagonal.

```python
def QRGramSchmidt(A):
    M = len(A)
    Q = np.zeros([M,M])
    R = np.zeros([M,M])
    
    for i in range(M):
        v = A[:,i] - np.dot(Q[:,:i] , np.dot(A[:,i], Q[:,:i]) )
        vv = np.linalg.norm(v) 
        Q[:,i] = v/vv
        R[i, i:] = np.dot(Q[:,i],A[:,i:])
    return Q, R


A = Sist.build_matrix_dynamic(1)
eps = 1
count = 0
while(eps > 1E-7 and count < 10000):
    Q, R = QRGramSchmidt(A)
    A = R @ Q
    eps = np.linalg.norm(A - np.diag(np.diag(A)))
    count += 1

print("\nNo caso da matriz simétrica: \n")
np.set_printoptions(precision=2,suppress=True)
print(A)

A = Sist.build_matrix_dynamic(5)
eps = 1
count = 0
while(eps > 1E-7 and count < 10000):
    Q, R = QRGramSchmidt(A)
    A = R @ Q
    eps = np.linalg.norm(A - np.diag(np.diag(A)))
    count += 1

print("\nNo caso da matriz não simétrica: \n")
np.set_printoptions(precision=2,suppress=True)
print(A)
```

6. É possível obter o os valores próprios de uma matriz pela determinação dos zeros do polinómio característico,
$$\text{det}\left(A-\lambda I\right)=0,$$
onde $I$ é a matriz identidade. Uma das formas de calcular o determinante de uma matriz é através da sua decomposição LU. O determinante de qualquer matriz triangular (superior ou inferior) é o produto dos elementos da diagonal. Compare os seus resultados com o das rotinas dos módulos do *Python*. 

Vamos usar a método da secante para encontrar os zeros do polinómio. Para evitar encontrar novamente os zeros já encontrados iremos dividir um polinómio que tem como zeros os zeros já conhecidos. Deste modo, no passo em que já foram encontrados $m$ zeros,
\begin{align}
f_m (\lambda) &= \frac{\text{det}\left(A-\lambda I\right) }{\prod_{i=1}^{m} (\lambda-\lambda_i)}
\end{align}
Na figura abaixo representamos $f_0(\lambda)$ ao passo que usamos a função $f_m(\lambda)$ para calcular os zeros.

```python
def LUDecomposition(A):
    """
    Escrever explicitamente as expressões
    """
    M = len(A)
    L = np.zeros([M,M],dtype=float)
    U = np.identity(M)
    
    for i in range(M):
        L[i:,i] = A[i:,i] - np.dot(L[i:,:i], U[:i,i])
        U[i, i+1:] = (A[i, i+1:] - np.dot(L[i,:i], U[:i, i+1:]) )/L[i,i]
    return L, U

def f(x, A, zeros):
    M = len(A)
    A1 = A - x * np.identity(M)
    # L, U, index, numberPermutations = LUDecomposition(A1)
    # value = np.prod(np.diag(L) * np.diag(U)) * (-1)**numberPermutations
    L, U = LUDecomposition(A1)
    value = np.prod(np.diag(L) * np.diag(U)) 
    if(len(zeros) > 0):
        value /= np.prod(x - zeros)  
    return value

M = 512
A = Sist.build_matrix_dynamic(5)
lambdas = np.linspace(-3.7,-0.02, M)
fv = np.zeros(M)
z1 = np.array([])
for i, l in enumerate(lambdas):
    fv[i] = f(l, A,z1)

def get_all_zeros(A):
    z1 = np.array([])
    for i in range(len(A)):
        x1, x2 = -5, 0.1
        f1, f2 = f(x1,A,z1), f(x2,A,z1)
        count = 0 
        while((np.abs(f2) > 1E-14) and count < 1000 ):
            x3 = (f2 * x1 - f1 * x2)/(f2 - f1)    
            x1 = x2
            f1 = f2
            x2 = x3
            f2 = f(x3,A,z1)
            count += 1
        z1 = np.concatenate((z1,np.array([x2])))    
    return np.sort(z1)

plt.figure(figsize=(15,5))
plt.plot(lambdas,fv )
plt.plot(lambdas,0*fv,'--')
z1 = get_all_zeros(A) 

for i in range(9):
    plt.vlines(z1[i],-1.1,5,'k',linestyles='dashed')
plt.ylabel(r'det$\left(A-\lambda I\right) $',fontsize=16)
tmp = plt.xlabel(r'$\lambda$',fontsize=16)
```

## Pergunta 2

**Evolução temporal do Sistema de Molas** - A evolução pode ser obtida a partir da diagonalização da matriz $A$. Contudo este problema de valores próprios é um pouco mais complicado uma vez que A não é simétrica. A matriz $A$ pode então ser diagonalizada à esquerda e à direita,
\begin{align}
A_{ij}r_{j}^{(\lambda)}	&= \lambda r_{i}^{(\lambda)}	&l_{i}^{(\lambda)}A_{ij}	&=\lambda l_{j}^{(\lambda)},
\end{align}
onde $r_{j}^{(\lambda)}$ e $l_{i}^{(\lambda)}$ são respectivamente os vectores próprios à esquerda e à direita associados aos valores próprios $\lambda$. Este problema pode ser reescrito como, 
\begin{align}
A_{ij}r_{i\alpha}	&=r_{i\beta}D_{\beta\alpha}	&l_{\alpha i}A_{ij}	&=D_{\alpha\beta}l_{\beta j} 
\end{align}
com $D_{\alpha\beta}=\lambda_{\alpha}\delta_{\alpha\beta}$  a matriz diagonal com os valores próprios e organizamos todos os vectores próprios em matrizes onde usamos os índices do alfabeto grego para indicar o respectivo vector próprio. Estas matrizes representam uma base para este espaço. Agora os vectores próprios à direita (e à esquerda) não são ortogonais entre si. A relação de ortogonalidade deverá ser dada por 
\begin{align}
l_{\beta i}r_{i\alpha}	&=\delta_{\alpha\beta}	&r_{i\alpha}l_{\alpha j}	&=\delta_{ij}
\end{align}
1. Para calcular os vectores próprios deste sistema podemos recordar que o vector próprio é solução da equação
   \begin{align}
   \left(A_{ij}-\lambda\delta_{ij}\right)x_{j}=0.
   \end{align}
   Será que consegue, sabendo os valores próprios obter os vectores usando as técnicas ensinadas no curso? Note que para um valor próprio, este sistema de equações torna-se indeterminado. Compare e comente os resultados que obteve para todos os $r_{\alpha}^{(\lambda)}$ e $l_{\alpha}^{(\lambda)}$ com os obtidos nos módulos *Python*.

Quando resolvemos com pivotagem o sistema de equações para um valor próprio não degenerado iremos obter no último passo o coeficiente nulo na diagonal. Este coeficiente significa que a última coordenada é indeterminada. Por isso podemos estipular um valor e a partir daí resolver o sistema de equações para as restantes coordenadas. 

Para matrizes simétricas a normalização é dada pelo produto dos vectores próprios à esquerda com os vectores próprios à direita. Para equilibrar estes vectores, normalizamos cada um de modo a que $l \cdot l^T = 1$ e $r^T \cdot r$ e por fim dividimos a o vector à direita e o vector à esquerda pela raíz quadrada do produto.


```python
def solveGaussJordanPivotagemParcial(A, b):
    M = len(b)
    linhas = np.arange(M)
    AA = np.zeros([M, M+1])
    AA[:M,:M] = A
    AA[:,M] = b
    
    for p in range(M-1): # depois desta iteração a 
        index =  p + np.argmax(np.abs(AA[linhas[p:], p]))
        linhas[p], linhas[index] = linhas[index], linhas[p]
        
        AA[linhas[p], : ] /= AA[linhas[p],p]

        for i in range(p):
            AA[linhas[i],p:] -= AA[linhas[p],p:] * AA[linhas[i],p]
        for i in range(p + 1, M):
            AA[linhas[i],p:] -= AA[linhas[p],p:] * AA[linhas[i],p]

    AA[:,-1] = - AA[linhas, -2]
    AA[-1,-1] = 1
    return AA[:,-1] / (AA[:,-1] @ AA[:,-1])



A1 = Sist.build_matrix_dynamic(5.0)
valoresProprios = get_all_zeros(A1)
vr = np.zeros([2*n + 1, 2*n + 1])
vl = np.zeros([2*n + 1, 2*n + 1])
for i in range(2*n + 1):
    A2 = A1 - valoresProprios[i]*np.identity(2*n + 1)
    xvr = solveGaussJordanPivotagemParcial(A2, np.zeros(2*n +1))
    xvl = solveGaussJordanPivotagemParcial(A2.T, np.zeros(2*n +1))
    norm = (xvl @ xvr)
    xvr /= np.sqrt(norm)
    xvl /= np.sqrt(norm)
    vr[:,i] = xvr[:]
    vl[:,i] = xvl[:]

np.set_printoptions(precision=2, suppress=True)
print("Os vectores próprios à esquerda e à direta são ortogonais entre si \n\n")
print(vl.T @ vr)
print("\n\nA matriz pode ser escrita a partir  dos seus valores próprios e vectores próprios\n\n")
print(A1 - vr@np.diag(valoresProprios)@vl.T)

print("comparando com o Python:")
w, l, r = sci.linalg.eig(A1,left=True)
print("\n os valores próprios têm que ser ordenados... \n")
index = np.argsort(np.real(w))
print(np.real(w[index]) )
print(valoresProprios)

print("\n Se o quociente das coordenadas dos vectores próprios à esquerda\n",
        "obtido pelo nosso método com as corrdenadas dos vectores obtidos com\n"
         "os métodos do scipy forem constantes significa que os vectores são paralelos\n\n"
    )

print(l[:,index]/vl)
print("\nO mesmo aplica-se aos vectores à direita ... \n")
print(r[:,index]/vr)
```

2. A evolução temporal deste sistema é dada por
   $$\frac{\partial^{2}x_{i}}{\partial t^{2}}=A_{ij}x_{j}$$
   Fazendo a mudança de variáveis, $y_{\beta}=l_{\beta i}x_{i}$ (logo $x_{i}=r_{i\beta}y_{\beta})$, obtemos,
   \begin{align}
   \frac{\partial^{2}y_{\alpha}}{\partial t^{2}}	&=l_{\beta i}A_{ij}r_{j\alpha}y_{\alpha}=D_{\beta\gamma}l_{\gamma j}r_{j\alpha}y_{\alpha} \\
	&=D_{\beta\gamma}\delta_{\gamma\alpha}y_{\alpha} \\
	&=\lambda_{\alpha}y_{\alpha}
\end{align}
um sistema de equações desacopladas. Para este sistema de equações, a solução mais geral é dada por,
$$y_{\alpha}(t)=A_{\alpha}e^{\sqrt{\lambda_{\alpha}}t}+B_{\alpha}e^{-\sqrt{\lambda_{\alpha}}t}.$$
Criando como condição inicial um conjunto de posições aleatorias geradas entre $x_{i}(0)\in]-0.01,0.01[$ e uma velocidade inicial nula para todas as massas. Calcule as condições iniciais $y_{\alpha}(0)$.

A partir da matriz podemos calcular os valores próprios e os vectores próprios à esquerda e à direita devidamente normalizados.
Projectando a condição inicial com os vectores próprios à esquerda podemos obter a condição inicial para $y_\alpha(0)$.
 Note-se que os valores próprios são todos negativos, o que implica que as raízes quadradas são números imaginários puros e a solução será oscilante.  

```python
A1 = Sist.build_matrix_dynamic(5.0)
w, l, r = sci.linalg.eig(A1,left=True)
norm = np.sqrt(np.diag(l.T@r))
for i in range(Sist.size):
    l[:,i] /= norm[i]
    r[:,i] /= norm[i]
x0 = 0.01*(2*rd.random_sample(Sist.size) - 1)
y0 = l.T @ x0
A0 = y0/2
B0 = A0
print(w)
```

```python
Sist64 = Molas(64)
for M in [1.0, 1024]:
    A1 = Sist64.build_matrix_dynamic(M)
    w, l, r = sci.linalg.eig(A1,left=True)
    norm = np.sqrt(np.diag(l.T@r))
    for i in range(Sist64.size):
        l[:,i] /= norm[i]
        r[:,i] /= norm[i]
    x0 = np.zeros(Sist64.size)
    x0[0] = 1/64
    y0 = l.T @ x0
    A0 = y0/2
    B0 = A0
    
    
    A1, b = Sist64.build_matrix_equilibrium(0.0)
    Xeq = solveGaussPivotagemTotal(A1,b)
    MT = 512
    tempos = np.linspace(0, 256, MT)
    yt = np.real((A0*np.exp(np.outer(np.sqrt(w),tempos)).T + B0* np.exp(np.outer(-np.sqrt(w),tempos)).T).T)
    xt = r @ yt
    plt.figure(figsize=(15,4))
    for i in range(Sist64.size):
        plt.plot(tempos, Xeq[i] + xt[i,:],'k', linewidth=0.75)
    tmp = plt.xlabel(r'$t$',fontsize=16)
    tmp = plt.ylabel(r'$x_i(t)$',fontsize=16)
    plt.show()
```
